```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid", font_scale=1.1)
```

```python
df['fecha'] = pd.to_datetime(df['fecha'], format='%Y-%m-%d')
df['equipo_1'] = df['equipo_1'].astype('string')
df['equipo_2'] = df['equipo_2'].astype('string')
df['url'] = df['url'].astype('string')
df['match_id'] = df['match_id'].astype('string')
```

```python
cols = ['posesion_1', 'posesion_2', 'disparos_totales_1',
       'disparos_totales_2', 'disparos_fuera_1', 'disparos_fuera_2',
       'paraden_1', 'paraden_2', 'corners_1', 'corners_2', 'tiros_libres_1',
       'tiros_libres_2']

print(f"DataFrame shape before outlier removal: {df.shape}")

def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 2.5 * IQR
    upper_bound = Q3 + 2.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)], Q1, Q3, IQR
    
for col in cols:

    if col not in ['posesion_1', 'posesion_2']:
        continue

    before = len(df)
    df, Q1, Q3, IQR = remove_outliers_iqr(df, col)
    after = len(df)
    print(f"{col} - Q1: {Q1}, Q3: {Q3}, IQR: {IQR}, lower_bound: {Q1 - 2.5 * IQR}, upper_bound: {Q3 + 2.5 * IQR}")
    print(f"Outliers eliminados en {col}: {before - after}")
    print("-" * 40)
df = df[df['equipo_1'] != "error"]

print(f"DataFrame final shape after outlier removal: {df.shape}")
```

```python

# Recorremos de a pares (0-1, 2-3, etc.)
for i in range(0, len(cols), 2):
    subset = cols[i:i+2]   # Tomamos las dos columnas
    n = len(subset)
    
    # Concatenar valores de las columnas del par
    combined = pd.concat([df[col] for col in subset], ignore_index=True)
    
    # Crear histograma conjunto
    plt.figure(figsize=(8, 5))
    sns.histplot(combined, bins=20, kde=True, color='skyblue')
    
    plt.title(f"Distribución conjunta de {', '.join(subset)}")
    plt.xlabel("Valor")
    plt.ylabel("Frecuencia")
    plt.tight_layout()
    plt.show()
```

```python
# construccion del grafo

# transformacion df a estructura de grafo
# in > out

columns = ['url','comp','stat','value_in', 'value_out', 'equipo_in', 'equipo_out']

# df para el grafo
graph_df = []

# iteramos sobre las filas del df y construimos el grafo
for index, row in df.iterrows():

    equipo_1 = row['equipo_1']
    equipo_2 = row['equipo_2']
    url = row['url']
    comp = 'liga'

    for col in (x for x in df.columns if x.endswith('_1')):
        col = col[:-2]  # removemos el sufijo _1
        if col == 'equipo' or col == 'paraden':
            continue

        if row[col+'_1'] > row[col+'_2']:
            value_in = row[col+'_1']
            value_out = row[col+'_2']
            equipo_in = equipo_1
            equipo_out = equipo_2

        elif row[col+'_1'] < row[col+'_2']:
            value_in = row[col+'_2']
            value_out = row[col+'_1']
            equipo_in = equipo_2
            equipo_out = equipo_1
        
        else:
            graph_df.append({
                'url': url,
                'comp': comp,
                'stat': col,
                'value_in': row[col+'_1'],
                'value_out': row[col+'_2'],
                'equipo_in': equipo_1,
                'equipo_out': equipo_2
            })

            graph_df.append({
                'url': url,
                'comp': comp,
                'stat': col,
                'value_in': row[col+'_2'],
                'value_out': row[col+'_1'],
                'equipo_in': equipo_2,
                'equipo_out': equipo_1
            })

        graph_df.append({
                'url': url,
                'comp': comp,
                'stat': col,
                'value_in': value_in,
                'value_out': value_out,
                'equipo_in': equipo_in,
                'equipo_out': equipo_out
            })

graph_df = pd.DataFrame(graph_df, columns=columns)
graph_df.head()
```
