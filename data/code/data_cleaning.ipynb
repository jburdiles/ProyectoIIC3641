```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid", font_scale=1.1)
```

```python
path = r"..\tables\output.csv"
df = pd.read_csv(path)
df.head()
```

```python
print(df.shape)
```

```python
df['fecha'] = pd.to_datetime(df['fecha'], format='%Y-%m-%d')
df['equipo_1'] = df['equipo_1'].astype('string')
df['equipo_2'] = df['equipo_2'].astype('string')
df['url'] = df['url'].astype('string')
df['match_id'] = df['match_id'].astype('string')
```

```python
df.describe()
```

```python
all_teams = pd.concat([df['equipo_1'], df['equipo_2']], axis=0)
all_teams.value_counts()
```

```python
cols = ['posesion_1', 'posesion_2', 'disparos_totales_1',
       'disparos_totales_2', 'disparos_fuera_1', 'disparos_fuera_2',
       'paraden_1', 'paraden_2', 'corners_1', 'corners_2', 'tiros_libres_1',
       'tiros_libres_2']

def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    
for col in cols:
    before = len(df)
    df = remove_outliers_iqr(df, col)
    after = len(df)
    print(f"Outliers eliminados en {col}: {before - after}")

df = df[df['equipo_1'] != "error"]
```

```python
df.describe()
```

```python
corr = df[cols].corr()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Matriz de correlación entre variables")
plt.show()
```

```python

# Recorremos de a pares (0-1, 2-3, etc.)
for i in range(0, len(cols), 2):
    subset = cols[i:i+2]   # Tomamos las dos columnas
    n = len(subset)
    
    # Concatenar valores de las columnas del par
    combined = pd.concat([df[col] for col in subset], ignore_index=True)
    
    # Crear histograma conjunto
    plt.figure(figsize=(8, 5))
    sns.histplot(combined, bins=20, kde=True, color='skyblue')
    
    plt.title(f"Distribución conjunta de {', '.join(subset)}")
    plt.xlabel("Valor")
    plt.ylabel("Frecuencia")
    plt.tight_layout()
    plt.show()
```

```python
# Recorremos de a pares (0-1, 2-3, etc.)
for i in range(0, len(cols), 2):
    subset = cols[i:i+2]
    n = len(subset)
    
    fig, axes = plt.subplots(1, n, figsize=(6*n, 4))
    
    # Si solo hay una columna (por si el número total es impar)
    if n == 1:
        axes = [axes]
    
    for ax, col in zip(axes, subset):
        sns.histplot(df[col], bins=20, kde=True, ax=ax)
        ax.set_title(f"Distribución de {col}")
    
    plt.tight_layout()
    plt.show()
```
